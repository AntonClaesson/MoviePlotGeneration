```python
import pandas as pd
import ast
```

```python
movie_data = pd.read_csv('MovieSummaries/movie.metadata.tsv', sep='\t', header=None)
```

```python
movie_data.rename(columns={0:"Wiki_ID", 2:"Title", 8:"Genre"}, inplace=True) 
movie_data.head()
```

```python
movie_info = movie_data[["Wiki_ID", "Title", "Genre"]]
for row in range(movie_info.shape[0]):

    row_genre = movie_info.loc[row, "Genre"]

    genre = ','.join(list(ast.literal_eval(row_genre).values()))

    movie_info.loc[row, "Genre"] = genre
```

```python
movie_info.head()
```

```python
movie_info.to_csv("movie_title_genre.tsv", sep="\t")
```

```python
movie_plot = pd.read_csv('MovieSummaries/plot_summaries.txt', sep='\t', header=None)
```

```python
movie_plot.rename(columns={0:"Wiki_ID", 1:"Plot"}, inplace=True)
```

```python
movie_info_plot = movie_info.merge(movie_plot, on="Wiki_ID")
```

```python
movie_info_plot
```

```python
with open('movie_info_plot_trainig.txt', 'w') as train_file, open('movie_info_plot_testing.txt', 'w') as test_file:
    for row in range(movie_info_plot.shape[0]):
        line = "<BOS>" + " "
        for g in movie_info_plot.loc[row, "Genre"].split(','):
            line += "<" + g+ "> "
        line += "<SEP> " + movie_info_plot.loc[row, "Plot"] + "<EOS>\n"
        if row < movie_info_plot.shape[0] - 1000:
            train_file.write(line)
        else:
            test_file.write(line)
```
